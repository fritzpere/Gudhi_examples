```python
import gudhi as gd
p = gd.read_off(off_file = 'tore3D_1307.off')
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import random

```

```python
def transform_to_plot(sample): 
    '''Function that transforms the data in order to represent it in matplotlib'''
    return np.asarray([point[0] for point in sample]),np.asarray([point[1] for point in sample]),np.asarray([point[2] for point in sample])
```

```python
p2=transform_to_plot(p)
fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(p2[0],p2[1],p2[2],zdir='z')
pyplot.show()
```

```python
fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(p2[0],p2[1],p2[2],zdir='y')
pyplot.show()
```
