```python
import gudhi as gd
p = gd.read_off(off_file = 'tore3D_300.off')
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import random
import numpy as np


```

```python
def transform_to_plot(sample): 
    '''Function that transforms the data in order to represent it in matplotlib'''
    return np.asarray([point[0] for point in sample]),np.asarray([point[1] for point in sample]),np.asarray([point[2] for point in sample])
```

```python

plot_points=transform_to_plot(p)
fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(plot_points[0],plot_points[1],plot_points[2],zdir='z')
pyplot.show()
pyplot.savefig('sample_maller_torus_z.png')
```

```python
fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(plot_points[0],plot_points[1],plot_points[2],zdir='y')
pyplot.show()
pyplot.savefig('sample_maller_torus_y.png')
```

```python
Rips_complex_sample = gd.RipsComplex(points =p, max_edge_length=2)
```

```python
Rips_simplex_tree_sample = Rips_complex_sample.create_simplex_tree(max_dimension=3) 
```

```python
diag_Rips = Rips_simplex_tree_sample.persistence()
```

```python
gd.plot_persistence_diagram(diag_Rips,legend=True, max_plots=100)
pyplot.savefig('Torus_diagram.png')
```

```python
gd.plot_persistence_barcode(diag_Rips,max_barcodes=100, legend=True)
pyplot.savefig('Torus_barcodes.png')
```

```python
#We add noise to see the stability of the persistence
p_noise=p+np.random.uniform(0,0.1,np.shape(p))
Rips_complex_noise = gd.RipsComplex(points =p_noise  , max_edge_length=2)
```

```python
Rips_simplex_tree_noise = Rips_complex_noise.create_simplex_tree(max_dimension=3) 
```

```python
diag_Rips_noise = Rips_simplex_tree_noise.persistence()
```

```python
gd.plot_persistence_diagram(diag_Rips_noise,legend=True, max_plots=100)
pyplot.savefig('Torus_diagram_noise.png')
```

```python
gd.plot_persistence_barcode(diag_Rips,max_barcodes=100, legend=True)
pyplot.savefig('Torus_barcodes_noise.png')
```
